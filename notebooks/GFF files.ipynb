```python
def parse_gene(row):
    chrom = row[0]
    start = row[3]
    end = row[4]
    strand = row[5]
    meta = dict([x.split('=') for x in row[8].split(';')])
    
    return {
        'chrom': chrom,
        'start': start,
        'end': end,
        'strand': strand,
        'meta': meta
    }

genes = {}

for i,row in t.iterrows():
    gene = parse_gene(row)
    genes[gene['meta']['ID']] = gene
    
```
